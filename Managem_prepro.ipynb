```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Financial Statement Documents:


## Income Statement:


```python
income_statement=pd.read_excel(r'Financial Statement(MNG).xlsx',sheet_name='Income Statement')
income_statement=income_statement.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns:
income_statement.columns=income_statement.columns.str.replace('FY','',regex=False)
income_statement.head()
```

## Balance Sheet:


```python
balance_sheet=pd.read_excel(r'Financial Statement(MNG).xlsx',sheet_name='Balance Sheet')
balance_sheet=balance_sheet.iloc[2:,1:-1].reset_index(drop=True)

#Rename columns:
balance_sheet.columns=balance_sheet.columns.str.replace('FY','',regex=False)
balance_sheet.head()
```

## Cash Flow:


```python
cash_flow=pd.read_excel(r'Financial Statement(MNG).xlsx',sheet_name='Cash Flow')
cash_flow=cash_flow.iloc[2:,1:-1].reset_index(drop=True)

#Rename columns:
cash_flow.columns=cash_flow.columns.str.replace('FY','',regex=False)
cash_flow.head()
```

## Ratios:


```python
ratios=pd.read_excel(r'Financial Statement(MNG).xlsx',sheet_name='Ratios')
ratios=ratios.iloc[2:,1:-1].reset_index(drop=True)

#Rename columns:
ratios.columns=ratios.columns.str.replace('FY','',regex=False)
# Delete Current column:
ratios=ratios.drop(columns='Current')
ratios.head()
```

# Transpose Data:


## Income Statement Transpoe:


```python
# Transpose data:
inc_trsp=income_statement.transpose().reset_index(drop=False)

# set first row as columns head:
inc_trsp.columns=inc_trsp.iloc[0]

# drop First row:
inc_trsp=inc_trsp.drop(0)
inc_trsp
```

```python
# Converte Fiscal Year column to datetime type:
inc_trsp['Fiscal Year']=pd.to_datetime(inc_trsp['Fiscal Year'], errors='coerce')
# change  xxxx-01-01 to xxxx-12-31 :
inc_trsp['Fiscal Year']=inc_trsp['Fiscal Year'].dt.year.astype(str)+ '-12-31'
inc_trsp['Fiscal Year']=pd.to_datetime(inc_trsp['Fiscal Year'],errors='coerce')
inc_trsp
```

## Balance Sheet Transpose:


```python
# Transpose data:
bs_trsp=balance_sheet.transpose().reset_index(drop=False)

# set first row as columns head:
bs_trsp.columns=bs_trsp.iloc[0]

# drop First row:
bs_trsp=bs_trsp.drop(0)
bs_trsp
```

```python
# Converte Fiscal Year column to datetime type:
bs_trsp['Fiscal Year']=pd.to_datetime(bs_trsp['Fiscal Year'],errors='coerce')
# change  xxxx-01-01 to xxxx-12-31 :
bs_trsp['Fiscal Year']=bs_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
bs_trsp['Fiscal Year']=pd.to_datetime(bs_trsp['Fiscal Year'],errors='coerce')
bs_trsp
```

## Cash Flow Transpose:


```python
# Transpose data:
cf_trsp=cash_flow.transpose().reset_index(drop=False)

# set first row as columns head:
cf_trsp.columns=cf_trsp.iloc[0]

# drop First row:
cf_trsp=cf_trsp.drop(0)
cf_trsp
```

```python
# Converte Fiscal Year column to datetime type:
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'],errors='coerce')
# change  xxxx-01-01 to xxxx-12-31 :
cf_trsp['Fiscal Year']=cf_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'],errors='coerce')
cf_trsp
```

## Ratios Transpose:


```python
# Transpose data:
ratios_trsp=ratios.transpose().reset_index(drop=False)

# set first row as columns head:
ratios_trsp.columns=ratios_trsp.iloc[0]

# drop First row:
ratios_trsp=ratios_trsp.drop(0)
ratios_trsp
```

# Feature Engineering:


## Variables:


```python
# Overview:
market_cap= ratios_trsp['Market Capitalization'].str.replace(',','').astype(float)
revenue= inc_trsp['Revenue'].str.replace(',','').astype(float)
revenue_growth=inc_trsp['Revenue Growth (YoY)'].str.replace(',','').str.replace('%','').astype(float) /100
cogs=inc_trsp['Cost of Revenue'].str.replace(',','').astype(float)
gross_profit=inc_trsp['Gross Profit'].str.replace(',','').astype(float)
ebitda=inc_trsp['EBITDA'].str.replace(',','').astype(float)
ebitda_margin=inc_trsp['EBITDA Margin'].str.replace(',','').str.replace('%','').astype(float) /100
ebit=inc_trsp['Operating Income'].str.replace(',','').astype(float)
ebit_margin=inc_trsp['EBIT Margin'].str.replace(r'[%,]','',regex=True).astype(float) /100 
ebt=inc_trsp['Pretax Income'].str.replace(',','').astype(float)
net_income=inc_trsp['Net Income'].str.replace(',','').astype(float)
net_income_growth=inc_trsp['Net Income Growth'].str.replace(r'[%,]','',regex=True).replace('-',np.nan).astype(float) /100
earining_per_share=inc_trsp['EPS (Basic)'].str.replace(',','').astype(float)
free_cash_flow=inc_trsp['Free Cash Flow'].str.replace(',','').astype(float)
total_assets=bs_trsp['Total Assets'].str.replace(',','').astype(float)
total_liabilities=bs_trsp['Total Liabilities'].str.replace(',','').astype(float)
total_equity=bs_trsp['Shareholders\' Equity'].str.replace(',','').astype(float)
operating_cash_flow=cf_trsp['Operating Cash Flow'].str.replace(',','').astype(float)
investing_cash_flow=cf_trsp['Investing Cash Flow'].str.replace(',','').astype(float)
financing_cash_flow=cf_trsp['Financing Cash Flow'].str.replace(',','').astype(float)
net_cash_flow=cf_trsp['Net Cash Flow'].str.replace(',','').astype(float)
## Activity ratios:
receivables=bs_trsp['Receivables'].str.replace(',','').astype(float)
begining_receivables=receivables.shift(-1,fill_value=np.mean(receivables))
average_recievables=(receivables+begining_receivables)/2
inventory=bs_trsp['Inventory'].str.replace(',','').astype(float)
begining_inventory=inventory.shift(-1,fill_value=np.mean(inventory))
average_inventory=(inventory+begining_inventory)/2
payables=bs_trsp['Accounts Payable'].str.replace(',','').astype(float)
begining_paybles=payables.shift(-1,fill_value=np.mean(payables))
average_payables=(payables+begining_paybles)/2
working_capital=bs_trsp['Working Capital'].str.replace(',','').astype(float)
begining_working_capital=working_capital.shift(-1,fill_value=np.mean(working_capital))
average_working_capital=(working_capital+begining_working_capital)/2
begening_total_equity=total_equity.shift(-1,fill_value=np.mean(total_equity))
average_total_equity=(total_equity+begening_total_equity)/2
## Liquidity Ratios:
current_assets=bs_trsp['Total Current Assets'].str.replace(',','').astype(float)
current_liabilities=bs_trsp['Total Current Liabilities'].str.replace(',','').astype(float)
cash_and_equivalents=bs_trsp['Cash & Equivalents'].str.replace(',','').astype(float)
## Solvency Ratios:
total_debt=bs_trsp['Total Debt'].str.replace(',','').astype(float)
begining_total_assets=total_assets.shift(-1,fill_value=np.mean(total_assets))
average_total_asset=(total_assets+begining_total_assets)/2
interest_expenses=inc_trsp['Interest Expense'].str.replace(',','').astype(float)
## Profitability:
## Valuation Ratios:
price_to_earning=ratios_trsp['PE Ratio'].str.replace(',','').astype(float)
price_to_sale=ratios_trsp['PS Ratio'].str.replace(',','').astype(float)
price_to_book=ratios_trsp['PB Ratio'].str.replace(',','').astype(float)
dividend_yield=ratios_trsp['Dividend Yield'].str.replace(r'[%,]','',regex=True).replace('-',np.nan).astype(float)/100
ev_ebitda=ratios_trsp['EV/EBITDA Ratio'].str.replace(',','').astype(float)
ev_ebit=ratios_trsp['EV/EBIT Ratio'].str.replace(',','').astype(float)
entreprise_value=ratios_trsp['Enterprise Value'].str.replace(',','').astype(float)

```

## Overview:


```python
overview= bs_trsp[['Fiscal Year']].copy()
overview['Market Cap']= market_cap * 1000000
overview['Revenue']= revenue * 1000000
overview['Revenue Growth']=revenue_growth
overview['COGS']=cogs * 1000000
overview['Gross Profit']=gross_profit * 1000000
overview['EBITDA']=ebitda * 1000000
overview['EBITDA %']=ebitda_margin
overview['EBIT']=ebit * 1000000
overview['EBIT %']=ebit_margin
overview['EBT']=ebt * 1000000
overview['Net Income']= net_income* 1000000
overview['Net Income %']=net_income_growth
overview['EPS']=earining_per_share
overview['Free Cash Flow']=free_cash_flow* 1000000
overview['Total Assets']=total_assets* 1000000
overview['Total Liabilities']=total_liabilities* 1000000
overview['Total Equity']=total_equity* 1000000
overview['Operating Cash Flow']=operating_cash_flow* 1000000
overview['Investing Cash Flow']=investing_cash_flow* 1000000
overview['Financing Cash Flow']=financing_cash_flow* 1000000
overview['Net Cash Flow']=net_cash_flow* 1000000
overview
```

## Activity Ratios:


```python
activ_ratios=inc_trsp[['Fiscal Year']].copy()
activ_ratios['Receivables']=receivables* 1000000
activ_ratios['Receivables Trunover']=revenue/average_recievables
activ_ratios['Inventory']=inventory* 1000000
activ_ratios['Inventory Turnover']=cogs/average_inventory
activ_ratios['Payables']=payables* 1000000
activ_ratios['Payables Turnover']=cogs/average_payables
activ_ratios['DSO']=(average_recievables/revenue)*365
activ_ratios['DIO']=(average_inventory/cogs)*365
activ_ratios['DPO']=(average_payables/cogs)*365
activ_ratios['CCC']=activ_ratios['DSO']+activ_ratios['DIO']-activ_ratios['DPO']
activ_ratios['Working Capital']=working_capital* 1000000
activ_ratios['Working Capital Turnover']=revenue/average_working_capital
activ_ratios['Equity Turnover']=revenue/average_total_equity
activ_ratios
```

## Liquidity ratios:


```python
liqu_ratios=bs_trsp[['Fiscal Year']].copy()
liqu_ratios['Current Assets']=current_assets *1000000
liqu_ratios['Current Liabilities']=current_liabilities*1000000
liqu_ratios['Current Ratio']=current_assets/current_liabilities
liqu_ratios['Quick Ratio']=(current_assets-inventory)/current_liabilities
liqu_ratios['Cash Ratio']=cash_and_equivalents/current_liabilities
liqu_ratios['Operating Cash Flow Ratio']=operating_cash_flow/current_liabilities
liqu_ratios
```

## Solvency Ratios


```python
solv_ratios=inc_trsp[['Fiscal Year']].copy()
solv_ratios['Debt to Equity']=total_debt/total_equity
solv_ratios['Debt to Capital']=total_debt/(total_debt+total_equity)
solv_ratios['Debt to Assets']=total_debt/total_assets
solv_ratios['Financial leverage']=average_total_asset/average_total_equity
solv_ratios['Interest coverage']=ebit/interest_expenses
solv_ratios
```

## Profitability Ratios:


```python
prof_ratios=inc_trsp[['Fiscal Year']].copy()
prof_ratios['Gross Profit Margin']=gross_profit/revenue
prof_ratios['Operating Profit(EBIT) Margin']=ebit/revenue
prof_ratios['Pre-tax Margin']=ebt/revenue
prof_ratios['Net Profit Margin']=net_income/revenue
prof_ratios['ROA']=net_income/ average_total_asset
prof_ratios['ROI']=net_income/average_total_equity
prof_ratios['Return on Capital (ROCE)']=ebit/(total_assets-current_liabilities)


prof_ratios
```

## Valuation Ratios:


```python
val_ratios=inc_trsp[['Fiscal Year']].copy()
val_ratios['Entreprise Value']=entreprise_value*1000000
val_ratios['EPS']=earining_per_share
val_ratios['P/E']=price_to_earning
val_ratios['P/S']=price_to_sale
val_ratios['P/B']=price_to_book
val_ratios['Dividend Yield']=dividend_yield
val_ratios['EV/EBITDA']=ev_ebitda
val_ratios['EV/EBIT']=ev_ebit
val_ratios
```

```python
data={
    'Income Statement':income_statement,
    'Balance Sheet':balance_sheet,
    'Cash Flow Statement':cash_flow,
    'Ratios':ratios,
    'Income Statement trsp':inc_trsp,
    'Balance Sheet trsp':bs_trsp,
    'Cash Flow Statement trsp':cf_trsp,
    'Ratios trsp':ratios_trsp,
    'Overview':overview,
    'Activity Ratios':activ_ratios,
    'Liquidity Ratios':liqu_ratios,
    'Solvency Ratios':solv_ratios,
    'Profitability Ratios':prof_ratios,
    'Valuation Ratios':val_ratios

}

file_name='Financial Ratios (Managem).xlsx'

with pd.ExcelWriter(file_name,engine='xlsxwriter') as f:
    for name,df in data.items():
        df.to_excel(f,sheet_name=name,index=False)

print(f'Excel File {file_name} Saved successfully')
```
